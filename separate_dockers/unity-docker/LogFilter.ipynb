```python
#get path to downloads folder
import os

#targetFolder = os.path.join(os.path.join(os.environ['USERPROFILE']), 'Downloads')
targetFolder = os.getcwd()
```

```python
#Load file named DoAllTestsLogs.txt from desktop
logLines = open(targetFolder + "\\DoAllTestsLogs.txt", "r").readlines()
```

```python
logLinesFiltered = []
```

```python
#define phrases that will be filtered out
filterPhrases = ["ubuntu.com","Reading database","Preparing to unpack", "Unpacking", "Selecting previously unselected package", "Setting up",
                 "memorysetup-", "ERROR: Shader", "WARNING: Shader", "[ALLOC_TEMP", "Initial Block Size", "Current Block Size", "Peak Allocated Bytes","Overflow Count",
                 "[ALLOC_DEFAULT","Peak usage frame count:", "Requested Block Size", "Peak Block count", "Peak Allocated memory", 
                 "Peak Large allocation bytes", "Used Block Count","Peak Allocated bytes","Used Block count","Large Block size","IL2CPP Free after allocator was destroyed"]
```

```python
#select only lines that do not contain filter phrases
for line in logLines:
    if not any(phrase in line for phrase in filterPhrases):
        logLinesFiltered.append(line)
```

```python
#print stats
print("Total lines: " + str(len(logLines)))
print("Filtered lines: " + str(len(logLinesFiltered)))
print("Filtered lines percentage: " + str(round((len(logLinesFiltered)/len(logLines))*100, 2)) + "%")
```

```python
print("Filtered lines:")
for line in logLinesFiltered:
    print(line,end="")
```

```python
#write filtered lines to file
with open(targetFolder + "\\DoAllTestsLogsFiltered.txt", "w") as file:
    for line in logLinesFiltered:
        file.write(line)
```
